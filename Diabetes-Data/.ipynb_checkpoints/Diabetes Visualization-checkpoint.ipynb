# 1.- Verificando o intervalo de séries temporais

como pode ser visto abaixo, as séries temporais têm intervalos muito diferentes entre elas, por isso, escolhemos tratar cada "observação" (neste caso: os pacientes), como se fosse um conjunto de dados ideal.

na visualização gerada abaixo, as linhas verde e amarela são a média e a mediana, respectivamente; então, é possível chegar à conclusão definitiva de que os intervalos são muito diferentes uns dos outros

```python
time_span = np.zeros(N_SAMPLES)
id_x = 0
for df_ref in dataframe_container:
    temp__ = df_ref.Time.max() - df_ref.Time.min()
    time_span[id_x] = temp__.days
    id_x += 1

plt.figure(figsize=(15,4))
plt.grid()

plt.plot(time_span, "or", label='Days of Samples')
plt.plot(time_span, "--k")

plt.plot(np.ones(N_SAMPLES) * np.mean(time_span), "-g", label='Mean')
plt.plot(np.ones(N_SAMPLES) * np.median(time_span), "-y", label='Median')

plt.ylabel("Days")
plt.xlabel("Paciente")

legend = plt.legend(loc='upper left', shadow=True, fontsize='x-large')
legend.get_frame().set_facecolor('#def5f9')

plt.show()
```

Lembrando que esta materia acadêmica é orientada para visualização de dados, então não é obrigatório processá-los, mesmo quando necessário para garantir uma representação visual correta, portanto, se os pacientes tiverem menos de 100 dias de registro, pode ser considerado bom o suficiente para ter uma visualização "correta".

```python
id_to_use = np.argwhere(time_span <= 100).T.tolist()[0]

print("""
{:.4f} % dos pacientes a usar (muestras por dia <= 100)
{:.4f} % dos pacientes nao se usaram (muestras por dia > 100)
""".format(
    (len(id_to_use) / len(time_span)) * 100,
    ((len(time_span) - len(id_to_use)) / len(time_span)) * 100))
```

se essa abordagem não oferecer a melhor visualização, a quantidade de amostras de pacientes deixados pode ser "reduzida", mas se não, o processo de visualização permaneceria o mesmo.

É importante lembrar que, no caso de ser necessário, um pré-pré-processamento pequeno poderia ser aplicado

Temos 4 tipos principais de códigos, que são a ***quantidade da glicose, quantidade de exercício, quantidade de insulina (e tipo), e a quantidade de alimento ingerido***, além de outros códigos que não podem ser agrupados com eles, eles são: ***sintomas hipoglicêmicos e outros eventos não especificados***

Assim, neste caso, optamos por plotar apenas o grupo principal entre eles, o tipo de insulina e a medida. Todas as amostras têm índice de tempo, ou seja, todas as amostras são séries temporais, nesta consideração, todas as visualizações serão orientadas no análise dos dados considerando o tempo como um fator principal.

(o código abaixo gera amostras onde são agrupadas por seu tipo de grupo, listado acima)

```python
glucose_cod = {
58: "Pre-breakfast",
59: "Post-breakfast",
60: "Pre-lunch",
61: "Post-lunch",
62: "Pre-supper",
63: "Post-supper",
64: "Pre-snack"
}

exercise_cod = {
69: 'Typical',
70: 'More-than-usual',
71: 'Less-than-usual'}

insuline_cod = {
33: "Regular",
34: "NPH",
35: "UltraLente"}

meal_cod = {
66:	"Typica",
67:	"More-than-usual",
68:	"Less-than-usual"}

others_codes = {
48:	"Unspecified blood glucose measurement",
57:	"Unspecified blood glucose measurement",
65:	"Hypoglycemic symptoms",
72:	"Unspecified special event"
}
```

```python
def convert_codes_by_dict(input_df, dict_cod):
    output = pd.DataFrame()
    output["Time"] = input_df.Time

    for code, label in dict_cod.items():
        output[label] = [None] * len(input_df.index)

        temp_ = input_df[input_df.Code == code]
        output[label].iloc[temp_.index] = temp_.Value

    output.index = output.Time
    output.drop(columns="Time", inplace=True)
    return output
```

```python
list_of_dict = [glucose_cod, exercise_cod, insuline_cod, meal_cod, others_codes]

cases_plot_ready = []
for ii in id_to_use:
    temp = [None] * len(list_of_dict)
    for jj in range(len(list_of_dict)):
        carry = convert_codes_by_dict(dataframe_container[ii], list_of_dict[jj])
        carry.index = (carry.index - carry.index.min()) + pd.to_datetime('1980-01-01') 
        
        if carry.notna().sum().sum() == 0:
            pass
        else:
            temp[jj] = carry
    
    cases_plot_ready += [(ii, copy(temp))]
```

# 3.- Plots

Nesta parte, plotamos apenas alguns casos específicos, os outros não serão plotados porque são coleções de dados  orientadas para a mineração de dados, tendo muitos problemas em sua composição (não no mesmo intervalo de tempo, etc.).

Em cada paciente, utilizou-se uma técnica de visualização diferente, onde se mostram diferentes comportamentos.

```python
def sanitaze_df(df_insuline):
    alt = df_insuline.copy()
    alt = alt[alt.notna().sum(axis=1) > 0]
    return alt[alt.columns[alt.notna().sum(axis=0) > 0].tolist()]
```

## 3.1.- Paciente 1

```python
df = sanitaze_df(cases_plot_ready[0][1][2])

ref_idx_coll = []
ref_idx_mean = []
for week in  df.index.week.unique():
    ref_new_idx = df[df.index.week == week].index.min().date()
    ref_mean = df[df.index.week == week].mean().values
    
    ref_idx_coll += [ref_new_idx]
    ref_idx_mean += [ref_mean]

mean_by_week = pd.DataFrame(ref_idx_mean, columns=df.columns)
mean_by_week.index = ref_idx_coll
mean_by_week.index = pd.to_datetime(mean_by_week.index)

mean_by_week.iplot(kind='bar', xTitle='Week', yTitle='Mean Weekly')
```

Essa visualização representa a quantidade média de insulina ingerida por semana, onde a Insulina Regular representa uma insulina "normal", a mais próxima em composição com a insulina humana, e a insulina NPH  é um tipo de insulina que trabalha mais tempo, sendo em combinação com a insulina regular uma opção de emergência quando uma crise diabética é percebida

Pode ser visto no gráfico, que o paciente sofreu (talvez) dois pré-crise durante o registro de 14 semanas, onde o último pode ser considerado uma "pequena" crise

## 3.3.- Paciente 33

```python
df = sanitaze_df(cases_plot_ready[26][1][2])

total_at = np.arange(1, len(df.index) + 1, dtype=int)
alt_df = df.cumsum() / np.matrix([total_at, total_at]).T
alt_df.columns = ["Cumulative Mean Regualar", "Cumulative Mean NPH"]

for col_name in df.columns:
    temp = df[col_name]
    
    alt_df["Scaled [0-1]" + col_name] = ((temp - temp.min()) / (temp.max() - temp.min())).values

py.iplot([{
    'x': alt_df.index,
    'y': alt_df[col].dropna(axis=0),
    'name': col
}  for col in alt_df.columns], xTitle='Day', yTitle='Value')
```

Nessa variação do gráfico de dispersão, é possível ver a variação na média dos valores (quantidade de insulina por tipo).

O principal motivo para fazer esse tipo de variação reside no desejo de saber se o paciente está melhorando sua condição (reduzindo a quantidade de insulina) ou não.

Além dessa modificação, modificamos os valores orignais, aplicando sobre ela uma [0-1] normalização, facilitando a visualização, evitando superposições indesejáveis.

**Podemos concluir, com base no gráfico acima, que o paciente está melhorando sua condição.**

Explicando parte do código que fizemos, a [0-1] normalização foi feita após o cálculo da média cumulativa. Isso permite uma visualização correta da média cumulativa, mas ainda pode-se ver os dados originais como referência.

## 3.4.- Paciente 38

```python
df = sanitaze_df(cases_plot_ready[31][1][2])

total_at = np.arange(1, len(df.index) + 1, dtype=int)
alt_df = df.cumsum() / np.matrix([total_at, total_at]).T
alt_df.columns = ["Cumulative Mean Regualar", "Cumulative Mean NPH"]

for col_name in df.columns:
    temp = df[col_name]
    alt_df["Scaled [0-1]" + col_name] = 1 * ((temp - temp.min()) / (temp.max() - temp.min())).values
    
alt_df["Date"] = alt_df.index
alt_df.index = total_at.tolist()
```

```python
data = []

alt_idx = np.arange(1, len(alt_df.Date.unique()) + 1)
to_grads = alt_idx
to_grads = (to_grads - to_grads.min()) / (to_grads.max() - to_grads.min())
to_grads *= 360

to_grads = pd.DataFrame(to_grads, columns=["grad"], index=alt_idx)
to_grads["Date"] = alt_df.Date.unique()

for column in alt_df.columns[:-1]:
    current_data = alt_df[column].dropna()
    
    active_dates = alt_df.iloc[current_data.index - 1].Date
    
    angles_ = []
    for ref_date in active_dates:
        angles_ += [np.round(to_grads[to_grads.Date == ref_date].grad.values.tolist()[0],1)]
    
    data += [
        go.Scatterpolar(
        r = current_data.values.tolist(),
        theta = angles_,
        name=column
    )
            ]

layout = go.Layout(
    showlegend = True
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

Essa visualização é uma modificação do gráfico de dispersão de antes, mas com outros dados (mas, a finalidade e as métricas são as mesmas), o uso de uma visualização radial (eixos polares) poderia ser útil na busca de padrões cíclicos.

Como pode ser visto, neste caso específico, o paciente mostra um "melhoramento" em sua saúde, mas, no final do mês (**o ciclo completo represente o mês de janeiro**) sua saúde não melhora em contraste com os primeiros 3 ~ 4 dias do mês, além disso, sua saúde apresenta comportamento quase oscilatório, melhorando e diminuindo constantemente.

O rótulo de "melhoria" é feito com base no uso médio de Insulina regular

## 3.5.- Paciente 41 e Paciente 52

Esta Visualização de Séries Temporais ajuda na busca de padrões, não apenas nos dados brutos, mas também com técnicas mais poderosas.

Para converter os dados brutos em dados tipo onda usamos uma conversão Z-Score, onde a média dos dados se tornou 0 e os valores se tornaram fatores de desvio padrão, gerando um intervalo quantizado (pelo menos os principais fatores) de valores que variam de negativos a positivos.

Depois de obter os dados em sua *forma de onda*, aplicamos uma transformada de Fourier, para visualizar se os padrões são reais ou apenas artefatos na visualização.

Para entender o que significa uma transformada de Fourier, precisaremos de muitas linhas, mas, em síntese, a parte real da transformada de Fourier mostra as freqüências mais relevantes e a parte imaginaria dela representa o complemento das frecuências.

Então, se pudermos ver alguns pontos altos na parte real, os dados têm um padrão oscilatório, mesmo pode que seja suficiente para gerar uma visualização espacial da fase a partir dele.

### 3.5.1.- Paciente 41

```python
df = sanitaze_df(cases_plot_ready[34][1][2])

alt_df = df.copy()

for col_name in df.columns:
    temp = df[col_name]
    if temp.std() == 0.0:
        alt_df[col_name] = temp * 0.0
    else:
        alt_df[col_name] = ((temp - temp.mean()) / temp.std()).values
        
py.iplot([{
    'x': alt_df.index,
    'y': alt_df[col].dropna(axis=0),
    'name': col
}  for col in alt_df.columns], xTitle='Day', yTitle='Value')
```

```python

### Fourier

fourier_df = pd.DataFrame(index=np.arange(len(alt_df.index)))

for col_name in df.columns:
    fourier = np.fft.fft(alt_df[col_name].dropna().values)
    fourier_df[col_name + " Fourier Real"] = np.nan
    fourier_df[col_name + " Fourier Real"][:len(fourier.real)] = fourier.real
    
    fourier_df[col_name + " Fourier Imag"] = np.nan
    fourier_df[col_name + " Fourier Imag"][:len(fourier.imag)] = fourier.imag

py.iplot([{
    'x': fourier_df.index,
    'y': fourier_df[col].dropna(axis=0),
    'name': col
}  for col in fourier_df.columns], xTitle='Day', yTitle='Value')
```

Este paciente pode ter tomado muita insulina NPH, mas, porque ele não alterna nenhuma quantidade ou tempo quando tomado, seu gráfico posterior e sua representação de Fourier é apenas uma linha.

Além disso, a análise da "oscilação" dos Dados da Insulina Regular pode ser representada como duas frecuências principais, portanto, é altamente provável que ela seja representada como uma onda, não uma onda suave, mas aproximada.
