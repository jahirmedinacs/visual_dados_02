```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt

from pprint import pprint as pprint
from copy import copy as copy


import os
import sys

import re

N_SAMPLES = 70

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
index_vals = []
files_name = [None] * N_SAMPLES
for ii in os.listdir():
    if "_csv_" in ii:
        index_vals += [int(ii[-2:])]
        files_name[index_vals[-1] - 1] = ii

index_vals.sort()
```

```python
code_names = pd.read_csv("./code_names.csv", sep=";", header=0)
display(code_names)
```

```python
dataframe_container = []
for ii in range(N_SAMPLES):
    temp = pd.read_csv(files_name[ii],index_col=0, header=0, sep="\t", encoding="utf-8")
    
    temp.Time = pd.to_datetime(temp.Time)
    
#     series_temp = pd.Series(temp[["Value", "Code"]], index=temp.Time)
    
    dataframe_container += [temp]
```

```python
column_id_code = 33

subplot_format = [N_SAMPLES, 1]
plt.figure(figsize=(20, 5 * N_SAMPLES))

for ii in range(N_SAMPLES):
#     test = pd.DataFrame()
    test = dataframe_container[ii]
    
    test = test[test.Code == column_id_code]
    
    temp_index = test.Time.values
    
    if test.empty:
        pass
    else:
        test_data = test.Value.values.tolist()

        series_test = pd.Series(data=test_data, index=temp_index)

        len_sers = series_test.shape[0]
        rltv_idx = np.arange(len_sers) + 1

        plt.subplot(subplot_format[0], subplot_format[1], ii + 1)

        series_test.plot(grid=True, style='r*')
        # series_test.plot(grid=True, style='c--')

        (series_test.cumsum() / rltv_idx).plot(grid=True, style='g.')
        (series_test.cumsum() / rltv_idx).plot(grid=True, style='y--')

plt.savefig("the_big_plot")
```
